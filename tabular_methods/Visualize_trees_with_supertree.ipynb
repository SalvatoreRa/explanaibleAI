# Supertree

Supertree is a Python package created for interactive and user-friendly visualization of decision trees within Jupyter Notebooks, Jupyter Lab, Google Colab, and other environments that support HTML rendering. This tool allows you to not only display decision trees but also interact with them directly within your notebook.

[original project](https://github.com/mljar/supertree)

## Simple example

```python
from supertree import SuperTree
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
# simple test with the iris dataset
iris = load_iris()
X, y = iris.data, iris.target
model = DecisionTreeClassifier()
model.fit(X, y)
# supertree visualization
super_tree = SuperTree(model, X, y, iris.feature_names, iris.target_names)
super_tree.show_tree()
```

```python
# Example with XGBoost
import xgboost as xgb
from xgboost import plot_tree
import matplotlib.pyplot as plt
# Load the iris dataset
iris = load_iris()
X, y = iris.data, iris.target

# Initialize and fit the XGBoost model
model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')
model.fit(X, y)

super_tree = SuperTree(model, X, y, iris.feature_names, iris.target_names)
super_tree.show_tree()
```

```python
import lightgbm as lgb
from lightgbm import plot_tree
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the iris dataset
iris = load_iris()
X, y = iris.data, iris.target

# Initialize and fit the LightGBM model
model = lgb.LGBMClassifier()
model.fit(X, y)

# Visualize the tree with built-in function of lightgbm
ax = lgb.plot_tree(model, tree_index=0, figsize=(20, 10))
plt.show()
```

```python
# with supertree
super_tree = SuperTree(model, X, y, iris.feature_names, iris.target_names)
super_tree.show_tree()
```

```python
# Random forest example
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
X = diabetes.data
y = diabetes.target
model = RandomForestRegressor(n_estimators=100, max_depth=3, random_state=42)
model.fit(X, y)
super_tree = SuperTree(model,X, y)
# show tree with index 2 in your notebook
super_tree.show_tree(2)
```
